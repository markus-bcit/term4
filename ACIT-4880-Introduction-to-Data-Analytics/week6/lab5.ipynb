```python

le = LabelEncoder()
clean_housing = housing
clean_housing['CRIM'] = le.fit_transform(clean_housing['CRIM'])
clean_housing['ZN'] = le.fit_transform(clean_housing['ZN'])
clean_housing['INDUS'] = le.fit_transform(clean_housing['INDUS'])
clean_housing['CHAS'] = le.fit_transform(clean_housing['CHAS'])
clean_housing['NOX'] = le.fit_transform(clean_housing['NOX'])
clean_housing['RM'] = le.fit_transform(clean_housing['RM'])
clean_housing['AGE'] = le.fit_transform(clean_housing['AGE'])
clean_housing['DIS'] = le.fit_transform(clean_housing['DIS'])
clean_housing['RAD'] = le.fit_transform(clean_housing['RAD'])
clean_housing['TAX'] = le.fit_transform(clean_housing['TAX'])
clean_housing['PTRATIO'] = le.fit_transform(clean_housing['PTRATIO'])
clean_housing['B'] = le.fit_transform(clean_housing['B'])
clean_housing['LSTAT'] = le.fit_transform(clean_housing['LSTAT'])
clean_housing['MEDV'] = le.fit_transform(clean_housing['MEDV'])
clean_housing
```

```python
cor = round(clean_housing.corr(),1)
cor
```

```python
import seaborn as sns

plt.figure(figsize = (8,6))
sns.heatmap(cor, annot=True, cmap=plt.cm.CMRmap_r)
plt.show()
```

```python
def correlation(dataset, threshold1, threshold2):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold1 and abs(corr_matrix.iloc[i,j]) < threshold2:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
df = housing
corr_matrix = housing.corr()
highly_correlated = set()
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > 0.7:
            highly_correlated.add((corr_matrix.columns[i], corr_matrix.columns[j]))

# Select one feature from each highly correlated pair for removal
features_to_drop = set()
for feat1, feat2 in highly_correlated:
    if corr_matrix[feat1].mean() > corr_matrix[feat2].mean():
        features_to_drop.add(feat2)
    else:
        features_to_drop.add(feat1)

# Remove highly correlated features
df_reduced = df.drop(columns=features_to_drop)

# Discuss the rationale behind selecting specific features for removal and potential implications for model performance and interpretability.
print("Selected features to drop:", features_to_drop)
```

CRIM: Per capita crime rate by town
ZN: Proportion of residential land zoned for large lots (over 25,000 sq.ft.)
INDUS: Proportion of non-retail business acres per town
CHAS: Charles River dummy variable (1 if tract bounds river; 0 otherwise)
NOX: Nitric oxides concentration (parts per 10 million)
RM: Average number of rooms per dwelling
AGE: Proportion of owner-occupied units built prior to 1940
DIS: Weighted distances to five Boston employment centers
RAD: Index of accessibility to radial highways
TAX: Full-value property tax rate per $10,000
PTRATIO: Pupil-teacher ratio by town
B: 1000(Bk - 0.63)^2 where Bk is the proportion of Black people by town
LSTAT: Percentage of lower status of the population
MEDV: Median value of owner-occupied homes in $1000's
