```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

INDUS: Proportion of non-retail business acres per town
CHAS: Charles River dummy variable (1 if tract bounds river; 0 otherwise)
NOX: Nitric oxides concentration (parts per 10 million)
RM: Average number of rooms per dwelling
AGE: Proportion of owner-occupied units built prior to 1940
DIS: Weighted distances to five Boston employment centers
RAD: Index of accessibility to radial highways
TAX: Full-value property tax rate per $10,000
PTRATIO: Pupil-teacher ratio by town
B: 1000(Bk - 0.63)^2 where Bk is the proportion of Black people by town

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the churn dataset
churn_data = pd.read_csv("churn.csv")

# Check the first few rows of the dataset
print(churn_data.head())
```

b) Investigate using bar chart the churn variable and print the percentages (Fig. 1).

```python
plt.figure(figsize=(10, 6))
sns.countplot(x="Int'l Plan", hue='Churn?', data=churn_data)
plt.title('Churn by International Plan')
plt.xlabel('International Plan')
plt.ylabel('Count')

plt.figure(figsize=(10, 6))
sns.countplot(x="Int'l Plan", hue='Churn?', data=churn_data, palette='Set2', hue_order=['False.', 'True.'], 
              dodge=False, order=['no', 'yes'], saturation=0.75)
plt.title('Normalized Churn by International Plan')
plt.xlabel("Int'l Plan")
plt.ylabel('Normalized Count')
plt.legend(title='Churn')
```

Generate the contingency table (Fig. 4):

```python
contingency_table = pd.crosstab(churn_data["Int'l Plan"], churn_data['Churn?'])
print("Contingency Table:")
print(contingency_table)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Day Mins', y='Eve Mins', data=churn_data, hue='Churn?', palette='Set2')
plt.title('Day Minutes vs. Evening Minutes')
plt.xlabel('Day Minutes')
plt.ylabel('Evening Minutes')
plt.legend(title='Churn')
```

Bin the Customer Service calls variable into Low (< 4 calls) and high (>= 4) and get the
churn rate for Low and High customers service calls

```python
churn_data['Service Calls Category'] = pd.cut(churn_data['CustServ Calls'], bins=[0, 4, float('inf')], labels=['Low', 'High'])

churn_rate_by_calls = churn_data.groupby('Service Calls Category')['Churn?'].value_counts(normalize=True) * 100
print("Churn Rate by Customer Service Calls Category:")
print(churn_rate_by_calls)
```

Get the correlation plot for Day Minutes, Day Charge, Night Minutes, Night Charge

```python
correlation_variables = ['Day Mins', 'Day Charge', 'Night Mins', 'Night Charge']

correlation_matrix = churn_data[correlation_variables].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Plot for Day and Night Variables')
plt.show()
```
